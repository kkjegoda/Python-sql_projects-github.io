```python
import matplotlib.pyplot as plt
a=["Krima","Mehul","Aarti"]
b=[30000,60000,80000]
c=[50000,80000,90000]
plt.plot(a,b,marker="o",ls="--",color="green",label="year1")
plt.plot(a,c,marker="*",ls="-",color="red",label="year2")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["day1","day2","day3","day4"]
y=[300,400,500,450]
y1=[600,350,200,150]
plt.plot(x,y,marker="o", ls ="--" ,color="red",label="week1")
plt.plot(x,y1,marker="*", ls ="-" ,color="green",label="week2",alpha=0.3)
plt.legend()
plt.show()
```
