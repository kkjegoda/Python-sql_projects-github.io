```python
import matplotlib .pyplot as plt
y=[98,78,88,95]
x=["part1","part2","part3","part4"]
colors=["red","green","yellow","blue","orange"]
plt.bar(x,y,color=colors,edgecolor="black")
plt.xlabel("Parts of Harry Potter",fontsize=17)
plt.ylabel("Populatity",fontsize=17)
plt.title("Populatity of Different parts of  Harry potter",fontsize=20)
plt.savefig("bar.png",facecolor="pink",pad_inches=0.9,bbox_inches="tight")
plt.show()
```
