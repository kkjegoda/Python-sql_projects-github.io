```python
import matplotlib.pyplot as plt
brands=["oneplue","Apple","Nokia","Radmi"]
x=[22,33,44,56]
c=["red","violet","blue","pink"]
ex=[0,0,0,0.1]
plt.pie(x,labels=brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=180)
plt.show()
```
